```python
def hist_box_plotS(data,col,bins="auto"):
  fig,axis = plt.subplots(ncols=2,figsize=(11,3))

  plt.style.use("ggplot")
  sns.histplot(data=data,x=col,bins=bins,ax=axis[0],kde=True)
  sns.boxplot(data=data,x=col,ax=axis[1])
```

```python
for col in df.columns:
  hist_box_plotS(df,col)
```

```python
# Import some libraries to build model
from sklearn.preprocessing import StandardScaler # used to convert train and test data in range between -1 to 1
# scaling is done when we have some features with high value and some features values is in decimal to make it equal...

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,recall_score,precision_score,confusion_matrix,f1_score,classification_report
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
display(X,y)
```

### ***Modeling***

```python
clf = DecisionTreeClassifier()
clf
```

### ***Data Splitting***

```python
# Simply we here used stratified sampling to get y data in proportion but data is already in proportion:
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
# standardize the values:
scaler = StandardScaler()
scaler
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_train
```

```python
x_test = scaler.fit_transform(x_test)
```

```python
x_test
```

```python
# lets train the model:
clf.fit(x_train,y_train)
```

```python
# test:
y_pred = clf.predict(x_test)
y_pred
```

```python
y_test.values
```

```python
plt.figure(figsize=(20,18))
tree.plot_tree(clf,feature_names = df.columns,class_names = ["Mine","Rock"],filled = True);
```

```python
# accuracy:
accuracy_score(y_test,y_pred)*100
# accuracy of DecisionTreeClassifier model is average
```

```python
# lets use logistic regression model:
model = LogisticRegression()
model
```

```python
# training:
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)*100
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
y_test.values
```

```python
# accuracy of model:
accuracy_score(y_test,y_pred)*100
# accuracy of log reg model is good
```

```python
# Confusion_matrix:
con_mat = confusion_matrix(y_test,y_pred)
con_mat
```

```python
# implement confusion matrix in chart:
sns.heatmap(data=con_mat,annot=True,linewidth=0.7,linecolor="white",fmt=".2f",cmap="Greens")
print(f"Here we have Total 63 rocks and Mine in river\n")
print(f"Objective is while war when submarine going from ocean Enemy planted explosure in form of Mine So Model needs to predict iits Rock or Mine:\n")
print(f"OUt of 63 actually 27 are rocks and 36 Mine:\n")
print(f"Out of 27 rocks --> model predicts that 21 are Rocks --> this statements lies under True Positive Category\n")
print(f"Remaining 6 are actually rocks but model predicts that they are Mine --> Wrongly Negative --> False Negtaive\n")
print(f"Out of 36 Mine model predicts that 28 are Mine --> True Negative\n")
print(f"Remaining 8 are actually Mine but model predicts that they are Rock --> Wrongly Positive --> False Positive\n\n")
```

```python
print(classification_report(y_test,y_pred))
```

```python
# lets check the recall :
# Accuracy of Negative predictions done by model
recall_score(y_test,y_pred)*100
```

```python
# Lets chcek precision:
# Accuracy of positive predictions done by the model:
precision_score(y_test,y_pred)*100
```

```python
# f1_score:
# harmonic mean of precision and recall:
f1_score(y_test,y_pred)
```
