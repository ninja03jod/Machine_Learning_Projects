```python
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.DataFrame()
df
```

```python
# 1 page has 20 records = Total we have 457 pages with 15122 rows lets append them:
for i in range(1,458):
  response = requests.get("https://api.themoviedb.org/3/movie/top_rated?api_key=17f80ce06535a2aee0acef962096d7ef&language=en-US&page={}".format(i))
  df_temp = pd.DataFrame(response.json()['results'])[['id','title','release_date','overview','popularity','vote_average','vote_count']]
  df = df.append(df_temp,ignore_index=True)
```
